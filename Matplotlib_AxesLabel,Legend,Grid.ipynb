```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[45,48,50,52,44,46,45]
min_t=[35,30,36,38,34,32,39]
avg_t=[40,39,43,45,39,39,42]
```

```python
plt.xlabel('Date')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend()
plt.grid()
```
