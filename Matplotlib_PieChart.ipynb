```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
population=[100,250,300,200]
cities=['Multan','Lahore','Karachi','Faisalabad']
```

```python
plt.axis('equal')
plt.pie(population,labels=cities,autopct='%0.0f%%',explode=[0,0.1,0,0])

plt.savefig("piechart.jpeg",transparent=True)
plt.title('Cities Data')
plt.show()
```
