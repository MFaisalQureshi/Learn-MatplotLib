```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
marks=[80,70,65,85,75,73,82,84,78,68]
plt.hist(marks,bins=[65,70,75,80],rwidth=0.95)
```

```python
boys_marks=[80,70,65,85,75,73,82,84,78,68]
girls_marks=[85,75,68,83,77,76,86,88,80,70]
plt.title('Marks')
plt.xlabel('Marks range')
plt.ylabel('Students')
plt.hist([boys_marks,girls_marks],bins=[65,70,75,80],rwidth=0.95,color=['green','red'],label=['Boys','Girls'])
plt.legend()
```
